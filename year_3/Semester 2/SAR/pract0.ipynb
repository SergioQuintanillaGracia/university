```python
import string

def isPangram(word):
    return len(set(string.ascii_lowercase).difference(set(word.lower()))) == 0

isPangram("The quick brown fox jumps over the lazy dog")
```

```python
def isPerfectPangram(word):
    return isPangram(word) and len(word.replace(' ', '')) == len(string.ascii_lowercase)

isPerfectPangram("The quick brown fox jumps over the lazy dog"),\
isPerfectPangram("Mr Jock TV quiz PhD bags few lynx"),\
isPerfectPangram("abcdefghijklmnopqrstuvwxyz")
```

```python
def areAnagrams(a, b):
    def countLetterOccurrences(word):
        occurrences = {}
        valid_chars = string.ascii_lowercase

        for char in word:
            # Only consider valid characters (letters)
            char = char.lower()
            if char not in valid_chars:
                continue
            
            occurrences[char] = occurrences[char] + 1 if char in occurrences else 1
        
        return occurrences
    
    occurrencesA = countLetterOccurrences(a)
    occurrencesB = countLetterOccurrences(b)

    return occurrencesA == occurrencesB

areAnagrams("Life on mars", "alien forms"),\
areAnagrams("Life on marsa", "alien forms")
```

```python
def grep(file, text):
    with open(file, 'r') as f:
        return [line.rstrip() for line in f if text in line]

grep('test.txt', 'la')
```

```python
from collections import Counter

def charFreq(text):
    c = Counter(text)
    return dict(c)

charFreq('abcabcaaabc')
```

```python
def balancedChain(text):
    counter = 0

    for c in text:
        if c == '[':
            counter += 1
        elif c == ']':
            counter -= 1
            if counter < 0:
                print('NOT OK')
                return
    
    if counter != 0:
        print('NOT OK')
    
    print('OK')

balancedChain('[]')
balancedChain('[][]')
balancedChain('[[][]]')
balancedChain('][')
balancedChain('][][')
balancedChain('[]][[]')
balancedChain('[hola]')
balancedChain(']hola[')
```
